2014년 ~ 2016년 cctv 수를 더하고 2013년 이전 cctv수로 나누어 최근 3년간 cctv 증가율을 계산하여 추가하세요

```python
per_list = []
for i in range(25):
    per = (CCTV_Seoul.iat[i, 3] + CCTV_Seoul.iat[i, 4] + CCTV_Seoul.iat[i, 5]) / CCTV_Seoul.iat[i, 2]
    per_list.append(per) 
    
CCTV_Seoul["증가율"] = per_list
CCTV_Seoul.tail()
```

각 지역구별 전체 인구를 활용하여 각 지역구별 외국인비율과 고령자비율을 계산하여 추가하세요.

```python
pop_Seoul.head()
```

```python
r_for_ls = []
r_old_ls = []
for i in range(77):
    r_for = pop_Seoul.iat[i, 4] / pop_Seoul.iat[i, 3] * 100
    r_for_ls.append(r_for)
    
    r_old = pop_Seoul.iat[i, 5] / pop_Seoul.iat[i, 3] * 100
    r_old_ls.append(r_old)

pop_Seoul["외국인비율(%)"] = r_for_ls
pop_Seoul["고령자비율(%)"] = r_old_ls

pop_Seoul.tail()
```
